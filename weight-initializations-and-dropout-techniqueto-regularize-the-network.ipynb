```python
import numpy as np
from keras.datasets import mnist
from matplotlib import pyplot
import math
from keras.utils.np_utils import to_categorical
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(X):
    return np.maximum(0.0, X)

def tanh(x):
    return np.tanh(x)
```

```python
def GaussDistribution(samples, mu = 0, sigma = 0.1):
    data = np.random.normal(mu, sigma, samples)
    return data.reshape(samples,1)
```

```python
def generateWeightsWithZeroValues(layers_n = 10, hidden = 500):    
    temp =  np.array([np.zeros([hidden, 1]) for i in range(layers_n)])
    print("Standard Deviation of sample is % s " % (np.std(temp)))
    return temp
```

```python
def generateWeights(layers_n = 10, hidden = 500):    
    temp =   np.array([np.random.rand(hidden, 1) for i in range(layers_n)])
    print("Standard Deviation of sample is % s " % (np.std(temp)))
    return temp
```

```python
def generateWeightsGaussdistribution(layers_n = 10, hidden = 500):    
    temp =   np.array([GaussDistribution(hidden) for i in range(layers_n)])
    print("Standard Deviation of sample is % s " % (np.std(temp)))
    return temp
```

```python
def generateWeightsBigValues(layers_n = 10, hidden = 500):    
    temp =   np.array([GaussDistribution(hidden, sigma = 1.0) for i in range(layers_n)])
    print("Standard Deviation of sample is % s " % (np.std(temp)))
    return temp
```

```python
def generateWeightsHe(layers_n = 10, hidden = 500):    
    temp =   np.array([GaussDistribution(hidden, sigma = np.sqrt(2/hidden)) for i in range(layers_n)])
    print("Standard Deviation of sample is % s " % (np.std(temp)))
    return temp
```

```python
def generateWeightsXavier(layers_n = 10, hidden = 500, samples = 1000):    
    temp =   np.array([GaussDistribution(hidden, sigma = np.sqrt(1/hidden)) for i in range(layers_n)])
    print("Standard Deviation of sample is % s " % (np.std(temp)))
    return temp
```

```python
def forward_pass_net(w_inputs,
                 data,
                 act,
                 logs = False):
           
    H = np.matmul(w_inputs, data.T)
    if logs:
        print("H: " + str(H))     

    A = act(H)
    if logs:                
        print("A: " + str(A))
        
    return A
```

```python
def train_net(w_in, act, data_, layers_n=10):
    return  np.array([forward_pass_net(w_in[i], data_[i], act, logs = False) for i in range(layers_n)])
```

```python
def plot(A, title):
    n = 10
    fig, ax = plt.subplots(1, n, figsize=(20, 18))
    fig.suptitle(title, fontsize=16)
    axes = ax.flatten()
    for i,j in zip(range(n), axes):
        j.hist(A[i], bins = 30)
        j.set_title('Layer ' + str(i+1))
        
    plt.show()
```

```python
import numpy as np
from keras.datasets import mnist
from matplotlib import pyplot
import math
from keras.utils.np_utils import to_categorical
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.datasets import fetch_openml

x, y = fetch_openml('mnist_784', version=1, return_X_y=True)
```

```python
#x = (x/255).astype('float32')
y = to_categorical(y)
train_X, test_X , train_y, test_y = train_test_split(x, y, test_size=0.15, random_state=42)
print('X_train: ' + str(train_X.shape))
print('Y_train: ' + str(train_y.shape))
```

```python
def softmax(X):
    exps = np.exp(X - np.max(X))
    return exps/np.sum(exps,axis=0)

def d_softmax(x):
    exp_element=np.exp(x-x.max())
    return exp_element/np.sum(exp_element,axis=0)*(1-exp_element/np.sum(exp_element,axis=0))

def D_relu(x):
    x[x <= 0] = 0
    x[x > 0] = 1
    return x

def relu(X):
    return np.maximum(0.0, X) 

def L2(w1, w2, rate = 0.00001):
    w1 = np.array(w1)**2
    w2 = np.array(w2)**2
    return (rate * (np.sum(w1) + np.sum(w2)))/2

def sigmoid(x):
    return 1/(np.exp(-x)+1)    

def d_sigmoid(x):
    return (np.exp(-x))/((np.exp(-x)+1)**2)

def loss(predicted, target):
    return (predicted - target) ** 2

def d_loss(predicted, target):
    return predicted - target

def batchGenarator(data, target,  batchSize = 10):
    sample=np.random.randint(0,data.shape[0],size=(batchSize))
    x=data[sample]
    y=target[sample]
    return x, y

def evaluate(weightsInputLayer,
             weightsOutputLayer,
             data, target, alpha, logs = False):

    H, A, H_OUTPUT, A_OUTPUT, loss_entropy, _ = forward_pass(weightsInputLayer,
                                                          weightsOutputLayer,
                                                          data, target, logs = logs, isTrain = False)
    pred = np.argmax(A_OUTPUT, axis=1)
    Y = np.argmax(target, axis=1)
    accuracy = (pred == Y).mean()
    
    return accuracy

def getNumberForBatch(data, batchSize):
    all_batches = data.shape[0] // batchSize 
    if data.shape[0] % batchSize != 0:
        all_batches += 1    
        
    return all_batches

def L2_regu(weightsInputLayer, lambda_L2 = 0.0001, isL2=False):
    return weightsInputLayer * lambda_L2

def init(x,y):
    np.random.seed(42)
    layer=np.random.uniform(-1.,1.,size=(x,y))/np.sqrt(x*y)
    return layer.astype(np.float32)

def network_init(data, outputLayerNeurons = 3, hiddenLayerNeurons = 5):  
    np.random.seed(42)
    l1=init(data.shape[1],hiddenLayerNeurons)
    l2=init(hiddenLayerNeurons,outputLayerNeurons)
    return l1, l2

def forward_pass(weightsInputLayer,
                 weightsOutputLayer,
                 data, target,
                 isL2 = True, L2rate = 0.0001, logs = False, keep_probs = [0.5], isTrain = True):
    
    H = data.dot(weightsInputLayer)
    if logs:
        print("H: " + str(H.shape))
        
    A = relu(H)
    if logs:                
        print("A: " + str(A.shape))
    
    keep_prob = keep_probs[0]
    D = np.random.rand(A.shape[0], A.shape[1])
    
    if isTrain:
        D = (D < keep_prob).astype(int)
        A = np.multiply(D, A) 
        A = A/keep_prob    

    if logs:                
        print("D: " + str(D.shape))
        
    H_OUTPUT = A.dot(weightsOutputLayer)
    if logs:
        print("H OUTPUT: " + str(H_OUTPUT.shape))
            
    A_OUTPUT = softmax(H_OUTPUT)
    if logs:
        print("A OUTPUT: " + str(A_OUTPUT.shape))
        
    if isL2:
        L2_ = L2(weightsOutputLayer, weightsInputLayer, L2rate)
        
    loss_entropy = loss(A_OUTPUT, target)
    if logs:
        print("Loss: " + str(loss_entropy))
        
    return H, A, H_OUTPUT, A_OUTPUT, loss_entropy, D


def backward_pass(H, A,
                  H_OUTPUT,
                  A_OUTPUT,
                  loss,
                   weightsInputLayer,
                   weightsOutputLayer,
                  data, target, alpha,
                  opt, hyper_param, D, keep_probs,
                  logs = False,
                  isL2 = False,
                  lambda_L2 = 0.001, isTrain = True):
    
    REL_D = D_relu(H)
    if logs:
        print("REL_D: " + str(REL_D.shape))
        
    SOFT_D = d_softmax(H_OUTPUT)
    if logs:
        print("SOFT_D: " + str(SOFT_D.shape))
        
    LOSS_D = d_loss(A_OUTPUT, target)
    if logs:
        print("LOSS_D: " + str(LOSS_D.shape))   
        
    g_out, g_input = copute_gradient(target,
                          LOSS_D,
                          A,
                          A_OUTPUT,
                          REL_D,
                          SOFT_D,
                          weightsInputLayer,
                          weightsOutputLayer,
                          data, alpha, logs,
                           hyper_param, D, keep_probs, opt,
                           isL2 = isL2,
                           lambda_L2 = lambda_L2,isTrain = isTrain)
    
    return REL_D, SOFT_D, LOSS_D, g_out, g_input
                
def copute_gradient(target,
                          loss,
                          A,
                          A_OUTPUT,
                          REL_D,
                          SOFT_D,
                          weightsInputLayer,
                          weightsOutputLayer,
                          data, alpha, logs,
                          hyper_param, D,
                          keep_probs,
                          opt = 'sgd',
                          isL2 = False,
                          lambda_L2 = 0.001, isTrain = True):
    
    arrL2_IN = L2_regu(weightsInputLayer, lambda_L2)
    arrL2_Out = L2_regu(weightsOutputLayer, lambda_L2) 

    gradient = loss * SOFT_D
    gradient_out = (A.T @ gradient) 

    if isL2:
        gradient_out + arrL2_Out
  
    dA = np.multiply(D, REL_D)
    # rescale
    dA = dA/keep_probs[0]
    
    gradient_rel = (np.array(weightsOutputLayer).dot(gradient.T)).T * dA    
    gradient_input =  data.T @ gradient_rel 
    if isL2:
        gradient_input + arrL2_IN
        
    op1 = hyper_param[0]
    op2 = hyper_param[1]
    
    if opt == 'momentum':
        gradient_out, v1 =  momentum(gradient_out, alpha, op1[0], rho = 0.9)
        gradient_input, v2 =  momentum(gradient_input, alpha, op2[0], rho = 0.9)
        hyper_param[0][0] = v1
        hyper_param[1][0] = v2
        
    elif  (opt == 'adam'):
        gradient_out, first_moment, second_moment =  adam(gradient_out, alpha, op1[0], op1[1], b1 = 0.9, b2 = 0.99)
        gradient_input, first_moment1, second_moment1 =  adam(gradient_input, alpha, op2[0], op2[1], b1 = 0.9, b2 = 0.99)
        hyper_param = [[first_moment, second_moment], [first_moment1, second_moment1]]

        
    elif  opt == 'rmspro':
        gradient_out, ad = rmspro(gradient_out, alpha, op1[0], t = 0.5)
        gradient_input, ad1 = rmspro(gradient_input, alpha, op2[0], t = 0.5)
        hyper_param[0][0] = ad
        hyper_param[1][0] = ad1
        
    elif  opt == 'adaGrad':
        gradient_out, ad = adaGrad(gradient_out, alpha, op1[0], t = 0.5)
        gradient_input, ad1 = adaGrad(gradient_input, alpha, op2[0], t = 0.5)
        hyper_param[0][0] = ad
        hyper_param[1][0] = ad1    
        
    else :
        gradient_out = alpha * gradient_out
        gradient_input = alpha * gradient_input
        
    return gradient_out, gradient_input

###

def momentum(dx, alpha = 0.001, vx = 0, rho = 0.9):
    vx = rho * vx + dx
    x = alpha * vx
    return x, vx

def adam(dx, alpha, first_moment = 0, second_moment = 0, b1 = 0.9, b2 = 0.99):      
    first_moment = b1 * first_moment + (1 - b1) * dx
    second_moment = b2 * second_moment + (1 - b2) * dx * dx
    x = alpha * first_moment/(np.sqrt(second_moment) + 1e-7)
    return x, first_moment, second_moment  

def rmspro(dx, alpha, grad_squ = 0, t = 0.5):
    decay_rate = alpha - t * alpha
    grad_squ = decay_rate * grad_squ + (1 - decay_rate) * dx * dx
    x = alpha * dx / (np.sqrt(grad_squ) + 1e-7)
    return x, grad_squ

def adaGrad(dx, alpha, grad_squ = 0, t = 0.5):
    grad_squ += dx * dx
    x = alpha * dx / (np.sqrt(grad_squ) + 1e-7)
    return x, grad_squ

###
    
def train(data, target, batchSize, batches,
          weightsInputLayer, weightsOutputLayer, epochs, 
          logs, alpha = 0.001, opt = "rmspro", isL2 = False,
          lambda_L2 = 0.001, isTrain = True):
    
    loss_w, accuracies = [], []
    hyper_param = [[0, 0],[0, 0]]
    keep_probs = [.5]
    
    for epoch in range(epochs):
        if logs:
            print("\n")
            print("Epoch: " + str(epoch))

        data_batch, target_batch  =  batchGenarator(data.copy(), target.copy(),
                                                         batchSize)


        H, A, H_OUTPUT, A_OUTPUT, loss_E, D = forward_pass(weightsInputLayer.copy(),
                                                            weightsOutputLayer.copy(),
                                                            data_batch.copy(),
                                                            target_batch.copy(),                                                
                                                            logs = logs, keep_probs = keep_probs,
                                                           isTrain = isTrain)
        loss_w.append((loss_E.mean()))

        REL_D, SOFT_D, LOSS_D, g_out, g_input = backward_pass(H.copy(),
                                                  A.copy(),
                                                  H_OUTPUT.copy(),
                                                  A_OUTPUT.copy(),
                                                  loss_E,                                                   
                                                  weightsInputLayer.copy(),                    
                                                  weightsOutputLayer.copy(),
                                                  data_batch.copy(),
                                                  target_batch.copy(),
                                                  alpha , opt, hyper_param, D, keep_probs, logs = logs,
                                                  isL2 = isL2, lambda_L2 = lambda_L2,
                                                  isTrain = isTrain)
            
        weightsOutputLayer -= g_out
        if logs:
            print("Weights out: " + str(weightsOutputLayer.shape))
                
        weightsInputLayer -= g_input
        if logs:
            print("Weights in: " + str(weightsInputLayer.shape))
                
        category=np.argmax(A_OUTPUT ,axis=1)
        y_=np.argmax(target_batch,axis=1)
        accuracy=(category==y_).mean()
        accuracies.append(accuracy)
        
        if logs:
            print("Accurancy : " + str(accuracy))
            
    return weightsInputLayer, weightsOutputLayer, loss_w, accuracies
```
